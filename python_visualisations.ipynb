Vizualising Data

We will use matplotlib tools, this tool is widely used for data analysis jobs. If we want to produce very fancy-looking graphs for web, this is not the right choice.

We will be using matplotlib.pyplot module. In the simplest use, pyplot uses an internal state in which you build the visualisation step by step. Once we are done with it, we can save it using savefig() or display it with show()

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot()

#creates a line chart, years on x-axis and gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("years")
plt.show()
```

For making publication quality graphs, you need to focus more on the internal code; matplotlib is capable of making much more complex graphs

Bar Plots

A bar chart is a good choice when you want to show how some quantity varies among some discrete set of items

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
xs = [i for i, _ in enumerate(movies)]
plt.bar(xs,num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
plt.xticks([i for i,_ in enumerate(movies)],movies)
```

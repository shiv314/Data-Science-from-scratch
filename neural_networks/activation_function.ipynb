```python
import numpy as np
inputs = [[1,2,3,2.5],[2.,5.,-1.,2],[-1.5,2.7,3.3,-0.8]]
exp_values = np.exp(inputs)
probabilities = exp_values/np.sum(exp_values,axis=1,keepdims=True)
probabilities
```

```python
class ReLU:
    def forward(self, inputs):
        self.output = np.maximum(inputs,0)

class Softmax:
    def forward(self, inputs):
        self.exp_values = np.exp(inputs)
        self.output = self.exp_values/np.sum(self.exp_values,axis=1,keepdims=True)

class DenseLayer:
    def __init__(self, n_inputs, n_neurons):
        self.weights = np.array(np.random.randn(n_inputs,n_neurons))
        self.biases = np.array(np.random.randn(1,n_neurons))
    
    def forward(self,inputs):
        self.output = np.dot(inputs,self.weights) + self.biases
```

```python
import nnfs
from nnfs.datasets import spiral_data
X,y = spiral_data(samples=100,classes=3)
print(X.shape)
```

```python
dense1 = DenseLayer(2,4)
activation1 = ReLU()
dense2 = DenseLayer(4,3)
activation2 = Softmax()

dense1.forward(X)
activation1.forward(dense1.output)
dense2.forward(activation1.output)
activation2.forward(dense2.output)
activation2.output
```

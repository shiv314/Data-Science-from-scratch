Since a Neural Network consists of multiple layers, we will define a class to implement these different layers so that we don't have to write the same code again and again

### Dataset used

We will be using the spiral_data dataset from nnfs package. This dataset contains 2D points arranged in space randomly and we will try to model these points using a Neural Network.

```python
from nnfs.datasets import spiral_data
import numpy as np
import nnfs
nnfs.init()
import matplotlib.pyplot as plt
X, y = spiral_data(samples=100, classes=3)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

Now, let's try to see the color classification of the different points. There are 3 different classes here, so let's try to visualise these.

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap = 'brg')
plt.show()
```

The neural network is not aware of this and it will try to learn from the data to classify these points. As you can see, this is not a trivial problem. This is a complex space and it will take some time to learn this.

```python
import numpy as np
import nnfs
from nnfs.datasets import spiral_data
nnfs.init()

#Dense layer
class Dense_Layer:
    def __init__(self,no_input,no_neurons):
        self.weights = np.array(np.random.randn(no_input,no_neurons))
        self.bias = np.array(np.random.randn(1,no_neurons))
    
    def forward(self,input):
        self.output = np.dot(input,self.weights) + self.bias
    
```

```python
X,y = spiral_data(samples=100,classes=3)
layer_1 = Dense_Layer(2,4)
layer_1.forward(X)
print(X.shape)
print(layer_1.output, layer_1.output.shape)
```

This is awesome, right? Now, we can simply use this Dense_Layer class for the feedforward process. Here, if you notice, each row represents the output from each successive neuron. And therefore, you see an output of 300 rows and 4 columns where each column represents the output from each neuron.

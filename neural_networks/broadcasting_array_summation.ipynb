### SUM

```python
import numpy as np
A  = [[1,2,3],[4,5,6],[7,8,9]]
print(A)
```

```python
print(np.sum(A))
```

```python
print(np.sum(A,axis=0))
```

axis = 0 means that you are summing the rows, this looks like you are summing along the column but actually you are summing the different 'rows' altogether.

```python
print(np.sum(A,axis=1))
```

axis = 1 means that you are summing the columns, this looks like you are summing along the row but you are actually summing the different 'columns' altogether. This creates a lot of confusion and this is how you can remember it.

Currently, if you look at the dimensions of the resultant array, you will notice that it is not a 2D array, it's a 1D array.

```python
np.sum(A,axis=1).shape
```

Now, if you want to retain the shape of the array, you can use a parameter called as keepdims, if you assign keepdims to be true, you can get a 2D array properly.

```python
np.sum(A,axis=1,keepdims=True).shape
```

```python
m = np.max(A,axis=1)
m.shape
```

```python
A - m
```

As, you notice this is something we don't want. Therefore, we use the argument keepdims = True for subtracting the proper array
